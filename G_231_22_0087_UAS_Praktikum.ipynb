```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from google.colab import files
uploaded = files.upload()
```

```python
df = pd.read_csv("stock.csv")
df.head(13)
```

```python
# 2a_Purchase Price

plt.figure(figsize=(10,5))
plt.hist(df['Purchase Price'], bins=25, color='#a7ed78')
plt.title('Histogram Distribusi Purchase Price',pad=20 , fontsize=15, color='red')
plt.xlabel('Purcahase Price($)', fontsize = 10, color='#1a3878')
plt.ylabel('Frequency',fontsize = 10, color='#1a3878')

plt.show()
```

```python
# 2a_Current Price

plt.figure(figsize=(10,5))
plt.hist(df['Current Price'], bins=25, color='#a7ed78')
plt.title('Histogram Distribusi Current Price',pad=20 , fontsize=15, color='red')
plt.xlabel('Current Price ($)', fontsize = 10, color='#1a3878')
plt.ylabel('Frequency',fontsize = 10, color='#1a3878')

plt.show()
```

```python
# 2a_Gain/Loss

plt.figure(figsize=(10,5))
plt.hist(df['Gain/Loss'], bins=25, color='#a7ed78')
plt.title('Histogram Distribusi Gain/Loss',pad=20 , fontsize=15, color='red')
plt.xlabel('Gain/Loss ($)', fontsize = 12, color='#1a3745')
plt.ylabel('Frequency',fontsize = 12, color='#1a3745')

plt.show()
```
